```python
connections= file[1]
goal_node= file[-1]
```

```python
print(connections)
print(goal_node)
```

```python
connections= int(connections.strip())
```

```python
graph= {}
for r in range(2, connections + 2):
  node = file[r].strip().split()
  n1, n2 = node
  n1= int(n1)
  n2= int(n2)

  if n2 not in graph:
      graph[n1]= list()
      if n1 not in graph[n1]:
        graph[n1].append(n2)
```

```python
print(graph)
```

```python
#BFS
from queue import Queue
visited_list= {}
distance= {}
parent_node_list= {}
bfs_traversal_output= []
bfs_Queue = Queue()

for node in graph.keys():
    visited_list[node]= False
    parent_node_list[node]= None
    distance[node]= -1

source_node= 0
visited_list[source_node] = True
distance[source_node]= 0
bfs_Queue.put(source_node)

while not bfs_Queue.empty():
    u= bfs_Queue.get()
    bfs_traversal_output.append(u)

    for v in graph:
        if not visited_list:
          visited_list[v]= True
          parent_node_list[v]= u
          distance[v] = distance[u] + 1
          bfs_Queue.put(v)
```
